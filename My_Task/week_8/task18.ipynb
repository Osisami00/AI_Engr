## Data Cleaning

```python
# Checking missing Data

possums1.isna().sum()

# -from the below result, we can see that there are two columns with missing data
# -the first column with missing data is age with 2 data missing and column footlgth with one data missing
```

```python
# To fill up missing space with average number of the columns


# FIrst, to fill up missing data in columns AGE
possums1['age'] = possums1['age'].fillna(possums1['age'].mean())


# Secondly, let's fill up the missing data in col footlgth

possums1["footlgth"] = possums1["footlgth"].fillna(possums1["footlgth"].mean())
```

```python
# Let check if we still have missing data
possums1.isna().sum()
```

```python
# Now, we have no missing data because we filled the missing data with the average number of each colums with missing data
```

```python
# Checking for Duplicate Data

possums1.duplicated().sum()

# from result below, we have no duplicated data
```

```python
# Standardizing sex column by romoving the whitespaces


# Removing the whitespaces

possums1['sex'] = possums1['sex'].str.strip()
possums1['sex']
```

## Center of Measure

```python
# Lets zoom down on data that spell out the numerical features of Possums
num_features = [ 'age', 'head_length', 'skull_width', 'total_length', 'tail_length', 'foot_length', 'earconch', 'eye', 'chest', 'belly']

stat1 = pd.DataFrame({
    'Mean' : possums1[num_features].mean().round(2),
    'Median' : possums1[num_features].median().round(2),
    'Mode' : possums1[num_features].mode().iloc[0].round(2),    #in case there is multiple mode
    'Variance' : possums1[num_features].var().round(2),
    'Standard Deviation' : possums1[num_features].std().round(2),
})

stat1
```
