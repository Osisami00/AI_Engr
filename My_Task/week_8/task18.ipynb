```python
# 1- Finding mean of Possum

# a - Finding mean of column case
possums_case_mean = np.mean(possums1["case"])
possums_case_median = np.median(possums1["case"])
possums_case_mode = possums1['case'].mode()[0]
print("Data on Case (Observation Number)")
print(f"Mean: {possums_case_mean:.1f}")
print(f"Median: {possums_case_median:.1f}")
print(f"Mode:")
print(f"{possums_case_mode}cm\n")



# b- Finding mean of column Site
print("Data on The site number where the possum was trapped.")
possums_site_mean = np.mean(possums1["site"])
possums_site_median = np.median(possums1["site"])
possums_site_mode = possums1['site'].mode()
print(f"Mean: {possums_site_mean:.1f}")
print(f"Median: {possums_site_median:.1f}")
print(f"Mode:")
print(f"{possums_site_mode}cm\n")



# c- Finding mean of column Age
print("Data on Age")
possums_age_mean = np.nanmean(possums1["age"])
possums_age_median = np.nanmedian(possums1["age"])
possums_age_mode = possums1['age'].mode()
print(f"Mean: {possums_age_mean:.1f}")
print(f"Median: {possums_age_median:.1f}")
print(f"Mode:")
print(f"{possums_age_mode}cm\n")



# d- Finding mean of column Headleanght
print("Data on Head length, in mm.")
possums_hdlngth_mean = np.mean(possums1["hdlngth"])
possums_hdlngth_median = np.median(possums1["hdlngth"])
possums_hdlngth_mode = possums1['hdlngth'].mode()
print(f"Mean: {possums_hdlngth_mean:.1f}mm")
print(f"Median: {possums_hdlngth_median:.1f}mm")
print(f"Mode:")
print(f"{possums_hdlngth_mode}cm\n")


# e- Finding mean of Skullwidth
print("Data on Skull width, in mm.")
possums_skullw_mean = np.mean(possums1["skullw"])
possums_skullw_median = np.median(possums1["skullw"])
possums_skullw_mode = possums1['skullw'].mode()
print(f"Mean: {possums_skullw_mean:.1f}mm")
print(f"Median: {possums_skullw_median:.1f}mm")
print(f"Mode:")
print(f"{possums_skullw_mode}cm\n")



# f- Finding mean of Total length
print("Data on Total length, in cm.")
possums_totlngth_mean = np.mean(possums1["totlngth"])
possums_totlngth_median = np.median(possums1["totlngth"])
possums_totlngth_mode = possums1['totlngth'].mode()
print(f"Mean: {possums_totlngth_mean:.1f}cm")
print(f"Median: {possums_totlngth_median:.1f}cm")
print(f"Mode:")
print(f"{possums_totlngth_mode}cm\n")



# g- Finding mean of Tail length
print("Data on Tail length, in cm.")
possums_taill_mean = np.mean(possums1["taill"])
possums_taill_median = np.median(possums1["taill"])
possums_taill_mode = possums1['taill'].mode()
print(f"Mean: {possums_taill_mean:.1f}cm")
print(f"Median: {possums_taill_median:.1f}cm")
print(f"Mode:")
print(f"{possums_taill_mode}cm\n")


# h- Finding mean of Foot length
print("Data on foot length")
possums_footlgth_mean = np.mean(possums1["footlgth"])
possums_footlgth_median = np.median(possums1["footlgth"])
possums_footlgth_mode = possums1['footlgth'].mode()
print(f"Mean: {possums_footlgth_mean:.1f}cm")
print(f"Median: {possums_footlgth_median:.1f}cm")
print(f"Mode:")
print(f"{possums_footlgth_mode}cm\n")



#i - Finding mean of Ear-conch
print("Data on Ear-conch, in mm.")
possums_earconch_mean = np.mean(possums1["earconch"])
possums_earconch_median = np.median(possums1["earconch"])
possums_earconch_mode = possums1['earconch'].mode()
print(f"Mean: {possums_earconch_mean:.1f}mm")
print(f"Median: {possums_earconch_median:.1f}mm")
print(f"Mode:")
print(f"{possums_earconch_mode}cm\n")


#j - Finding mean of eye distance from medial canthus to lateral canthus of right eye
print("Data on Eye Distance from Medial Canthus, in mm.")
possums_eye_mean = np.mean(possums1["eye"])
possums_eye_median = np.median(possums1["eye"])
possums_eye_mode = possums1['eye'].mode()
print(f"Mean: {possums_eye_mean:.1f}mm")
print(f"Median: {possums_eye_median:.1f}mm")
print(f"Mode:")
print(f"{possums_eye_mode}cm\n")


#k - Finding mean of chest girth (in cm)
print("Data on Chest girth, in cm.")
possums_chest_mean = np.mean(possums1["chest"])
possums_chest_median = np.median(possums1["chest"])
possums_chest_mode = possums1['chest'].mode()
print(f"Mean: {possums_chest_mean:.1f}cm")
print(f"Median: {possums_chest_median:.1f}cm")
print(f"Mode:")
print(f"{possums_chest_mode}cm\n")


#l - Finding mean of belly girth (in cm)
print("Data on belly girth, in cm.")
possums_belly_mean = np.mean(possums1["belly"])
possums_belly_median = np.median(possums1["belly"])
possums_belly_mode = possums1['belly'].mode()
print(f"Mean: {possums_belly_mean:.1f}cm")
print(f"Median: {possums_belly_mean:.1f}cm")
print(f"Mode:")
print(f"{possums_belly_mode}cm\n")




# possums_mean = np.mean(possums1.select_dtypes(include=[np.number]))
# print("The mean of possum is:", possums_mean)


```
