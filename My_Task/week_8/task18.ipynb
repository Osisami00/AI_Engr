## Data Cleaning

```python
# Checking missing Data

possums1.isna().sum()

# -from the below result, we can see that there are two columns with missing data
# -the first column with missing data is age with 2 data missing and column footlgth with one data missing
```

```python
# To fill up missing space with average number of the columns


# FIrst, to fill up missing data in columns AGE
possums1['age'] = possums1['age'].fillna(possums1['age'].mean())


# Secondly, let's fill up the missing data in col footlgth

possums1["footlgth"] = possums1["footlgth"].fillna(possums1["footlgth"].mean())
```

```python
# Let check if we still have missing data
possums1.isna().sum()
```

```python
# Now, we have no missing data because we filled the missing data with the average number of each colums with missing data
```

# Measure of Spread

```python
# let us check the head again

possums1.head(5)
```

```python
# Checking the Measure of Spread

num_features = [ 'age', 'head_length', 'skull_width', 'total_length', 'tail_length', 'foot_length', 'earconch', 'eye', 'chest', 'belly']

stat3 = pd.DataFrame({
    "Max" : possums1[num_features].max().round(2),
    "Min" : possums1[num_features].min().round(2),
    "Range" : np.subtract(possums1[num_features].max(),  possums1[num_features].min()).round(2),
    "25th Percentile" : possums1[num_features].quantile(0.25).round(2),
    "50th Percentile" : possums1[num_features].quantile(0.50).round(2),
    "75th Percentile" : possums1[num_features].quantile(0.75).round(2),
    "IQR" : np.subtract(possums1[num_features].quantile(0.75), possums1[num_features].quantile(0.25)).round(2),
    "Skewness" : possums1[num_features].skew().round(2),
    "Kurtosis" : possums1[num_features].kurtosis().round()
})

stat3
```

Overall Insights
1- Age: the maximum age use in the data is 9yrs and min is 1yr with the range of 8.0 show that the data is spread out.the IQR of 0.54 shows the data is consistent and the kurtosis shows that the data is perfectly symmetry

2- head leangth highest data is 103.1 with minimum data is 82.5, thereby giving us the raange of 20.6. This range shows that the data is highly spread out and it is second highest compare to other features analyzed. The IQR is 4.05 ehich shows data is somewhat consistent but not as consistent as the age feature.The kurtosis od 1.0 shows that the data os mo


Features like eye diameter, skull width, chest circumference show low variation → these are stable biological traits.

Features like foot length, ear length, total length, belly circumference show higher variation → possibly influenced by age, sex, or ecological factors.
