```python
# CREATNG DATA FRAME

data = {
    'Name': ['Chris', 'Ayo', 'Chisom'],
    'Age': [26, 24, 22],
    'Home_town': ['Benin', 'Ibadan', 'Enugu']
}

# create DataFrame
df = pd.DataFrame(data, index=['i', 'ii', 'iii'])
df.head()
```

```python
# lets do the samething by using list of dict

data2 = [
    {'Name': 'Chris', 'Age': 26, 'Home_Town': 'Benin'},
    {'Name': 'Ayo', 'Age': 24, 'Home_Town': 'Ibadan'},
    {'Name': 'Chisom', 'Age': 22, 'Home_Town': 'Enugu'}
]

df2 = pd.DataFrame(data2)
df2
```

```python
# Lets do the same woth list of list
data2 = [
    {'Name': 'Chris', 'Age': 26, 'Home_Town': 'Benin'},
    {'Name': 'Ayo', 'Age': 24, 'Home_Town': 'Ibadan'},
    {'Name': 'Chisom', 'Age': 22, 'Home_Town': 'Enugu'}
]

df3 = pd.DataFrame(data2)
df3
```
